# Modeling

```python
from sklearn.model_selection import train_test_split 
```

```python
X = sb_class.drop(columns = ['person', 'next_purchase_range'])
y = sb_class['next_purchase_range']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_scaled_df = pd.DataFrame(X_train, columns = X.columns)
```

```python
X_scaled_df.head(10)
```

```python
X_scaled_df.mean()
```

```python
X_scaled_df.std()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
sns.set(rc = {'figure.figsize':(12,8)}, font_scale = 2)
clf_cmatrix = confusion_matrix(y_test, y_pred)
sns.heatmap(clf_cmatrix, annot=True, cmap = 'Blues', fmt = 'd').set(title = 'Confusion matrix with 3 classes');
```

```python
print("Classification report:")
print(classification_report(y_test, y_pred))
print("Accuracy score:")
print(accuracy_score(y_test, y_pred))
```

```python
y_test.value_counts()
```

# Checking percentiles
Let's see what the percentiles are for each 24-hour increment. We'll visualize with a line plot.

```python
from scipy import stats
```

```python
hours = [i for i in range(24, 678, 24)]
```

```python
percentiles = []
for hour in hours:
    percentiles.append(stats.percentileofscore(sb_class['next_purchase'], hour))
```

```python
hours_dict = {'hours': hours, 'percentiles': percentiles}
```

```python
hours_df = pd.DataFrame(hours_dict)
```

```python
sns.lineplot(x = 'hours', y = 'percentiles', data = hours_df)
plt.xticks([120, 240, 360, 480, 600])
plt.show()
```

# Let's try 9 classes
It looks like the percentile rises pretty sharply for the first 7 days, then levels off after that. So we can try to classify our customers by the day (or 24-hour interval) of their next purchase up to 7 days, then another bucket for 8-10 days, then the last one for more than 10 days.

```python
sb_class['next_purchase_range'] = 8 #within 1 day
sb_class.loc[sb_class.next_purchase>24,'next_purchase_range'] = 7 #within 2 days
sb_class.loc[sb_class.next_purchase>48,'next_purchase_range'] = 6 #within 3 days
sb_class.loc[sb_class.next_purchase>72,'next_purchase_range'] = 5 #within 4 days
sb_class.loc[sb_class.next_purchase>96,'next_purchase_range'] = 4 #within 5 days
sb_class.loc[sb_class.next_purchase>120,'next_purchase_range'] = 3 #within 6 days
sb_class.loc[sb_class.next_purchase>144,'next_purchase_range'] = 2 #within 7 days
sb_class.loc[sb_class.next_purchase>168,'next_purchase_range'] = 1 #within 8, 9 or 10 days
sb_class.loc[sb_class.next_purchase>240,'next_purchase_range'] = 0 #more than 10 days
```

And it looks like there's no extreme imbalance among the classes.

```python
sb_class['next_purchase_range'].value_counts()
```

# Reprocessing for new target variable
We'll train-test split and standardize again.

```python
Xlr = sb_class.drop(columns = ['person', 'next_purchase_range'])
ylr = sb_class['next_purchase_range']
```

```python
Xlr_train, Xlr_test, ylr_train, ylr_test = train_test_split(Xlr, ylr, test_size = 0.3, random_state = 42)
```

```python
Xlr_train = scaler.fit_transform(Xlr_train)
Xlr_test = scaler.transform(Xlr_test)
```

```python
Xlr_scaled_df = pd.DataFrame(Xlr_train, columns = Xlr.columns)
```

```python
Xlr_scaled_df.head(10)
```

```python
Xlr_scaled_df.mean()
```

```python
Xlr_scaled_df.std()
```

```python
clfr = LogisticRegression()
clfr.fit(Xlr_train, ylr_train)
ylr_pred = clfr.predict(Xlr_test)
clfr_cmatrix = confusion_matrix(ylr_test, ylr_pred)
```

```python
print("Classification report:")
print(classification_report(ylr_test, ylr_pred))
print("Accuracy score:")
print(accuracy_score(ylr_test, ylr_pred))
```

```python
sns.set(rc = {'figure.figsize':(12,8)}, font_scale = 2)
sns.heatmap(clfr_cmatrix, annot=True, cmap = 'Blues', fmt = 'd').set(title = 'Confusion matrix with 9 classes');
```

# Baseline model
Our baseline accuracy score is 82.88 percent. Definitely some room for improvement. Now we'll regularize the C parameter.

```python
#from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"C": [0.001, 0.1, 1, 10, 100]}
logreg = LogisticRegression()
logreg_cv = GridSearchCV(logreg, param_grid, cv=5)
logreg_cv.fit(Xlr_train, ylr_train) 
logreg_cv.best_params_
```

```python
logreg_cv.best_score_
```

```python
best_clsfr = LogisticRegression(C=100)
best_clsfr.fit(Xlr_train, ylr_train)
ylr_pred = best_clsfr.predict(Xlr_test)
best_clsfr_cmatrix = confusion_matrix(ylr_test, ylr_pred)
```

```python
print("Classification report:")
print(classification_report(ylr_test, ylr_pred))
print("Accuracy score:")
print(accuracy_score(ylr_test, ylr_pred))
```

```python
sns.set(rc = {'figure.figsize':(12,8)}, font_scale = 2)
sns.heatmap(best_clsfr_cmatrix, annot=True, cmap = 'Blues', fmt = 'd').set(title = 'Confusion matrix with regularization');
```
